```python
d1 = df['cities'].apply(lambda st : st.split(','))
x = zip(d1.index,d1.values)
st = []
for i,k in x:
#     print(i,k)
    for j in k:
#         print(j)
        e=(str(df.loc[i,'index'])+' '+j.replace(':',' ')).split(' ')
#         print(e)
        if len(e) == 4:
            e[1] = e[1]+ ' ' +e[2]
            e[2] = e[3]
#             print(e)
        st.append(e)
```

```python
df_cities_clean=pd.DataFrame(st)
```
