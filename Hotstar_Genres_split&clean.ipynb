```python
import pandas as pd
```

```python
df=pd.read_json('/home/pa1/Documents/Hackathon/Hotstar/train_data.json')
```

```python
df=df.T
```

```python
df.reset_index(inplace=True)
```

```python
d1 = df['genres'].apply(lambda st : st.split(','))
x = zip(d1.index,d1.values)
st = []
for i,k in x:
#     print(i,k)
    for j in k:
#         print(j)
        e=(str(df.loc[i,'index'])+' '+j.replace(':',' ')).split(' ')
#         print(e)
#         if len(e) == 4:
#             e[1] = e[1]+ ' ' +e[2]
#             e[2] = e[3]
# #             print(e)
        st.append(e)
```

```python
df_genres_clean=pd.DataFrame(st)
```

```python
total_cols=list(df_genres_clean[1].value_counts().index)
```

```python
sports =[ 'Football', 'Badminton', 'Sport', 'Hockey', 'Kabaddi',
       'Formula1', 'Tennis', 'Athletics', 'Table Tennis', 'FormulaE',
       'Volleyball', 'Boxing', 'Swimming', 'IndiaVsSa']
```

```python
final_genres=list(set(total_cols)-set(sports))
```
